```python
# 콤마가 있는 문자에서 콤마 없애기 방법1
temp = '12,523'
tempsplited = temp.split(',')
print(tempsplited)
print(''.join(tempsplited))
```

```python
# 분양가격 컬럼에 콤마(,)가 있는 경우 콤마(,)를 제거('5,123' -> 5123)
# 공백문자가 있는 경우 np.nan로 바꾸는 함수
def commadrop(row):
    # row = row.copy() # 워닝 출력시 카피를 할지에 대해 물어보는 내용으로 작성하면 워닝출력안함 
    pd.options.mode.copy_on_write = True # 데이터 복사를 최소화 (워닝 출력시)
    price = row['분양가격(제곱미터)']
    if row.notnull()['분양가격(제곱미터)']: # 분양가격이 결측치가 아닌 경우
        if price.find(',')!=-1: # 콤마(,) 없애기
#             row['제곱미터당분양가격'] = price.replace(',','')
            pricesplited = price.split(',')
            row['제곱미터당분양가격'] = ''.join(pricesplited)
        #   row['제곱미터당분양가격'] = ''.join(price.split(','))
        elif price.strip()=='':
            row['제곱미터당분양가격'] = np.nan
        else :
            row['제곱미터당분양가격'] = price
    else : # 분양가격이 결측치인 경우
        row['제곱미터당분양가격'] = np.nan
    return row

commadrop(df_last.loc[0]) # 콤마 제외
commadrop(df_last.loc[1]) # 원래 콤마 없는 데이터
commadrop(df_last.loc[28]) # 공백문자의 경우 nan
commadrop(df_last.loc[368]) # 원래 nan의 경우 nan
```

```python
# 공백문자 결측치로, 콤마는 제거하기를 모든 행에 적용
df_last = df_last.apply(commadrop, axis= 1)
df_last.head(3)
```

### map

```python
# '3,198' -> 3198 return, '3198' -> 3198 return, 공백문자 -> 결측치 return, 결측치 -> 결측치 return
def commadrop2(price):
    # np.isnan(값) : nan만 결측치로 인식하고 결측치인지 여부 return
    # pd.isna(값) , pd.isnull(값) : nan과 None을 모두 결측치로 인식하고 결측치인지 여부 return
    # 값 is np.nan : 값이 nan인지 여부 return
    if price is np.nan :
        return np.nan
    elif price.strip() == '':
          return np.nan
    elif price.find(',') != -1:
        pricesplited = price.split(',')
        price = ''.join(pricesplited)
        # price = ''.join(price.split(','))
        return int(price)
    else :
        return int(price)
commadrop2(df_last.loc[0, '분양가격(제곱미터)']) # 콤마없는 문자를 숫자로
commadrop2(df_last.loc[1, '분양가격(제곱미터)']) # 콤마제외한 문자를 숫자로
commadrop2(df_last.loc[28, '분양가격(제곱미터)']) # 공백문자를 nan
commadrop2(df_last.loc[368, '분양가격(제곱미터)']) # nan를 nan
```

- apply(함수, axis) : 1차원, 2차원(axis 지정. axis기본값0) 사용
- map(함수) : 1차원에서만 사용
- applymap(함수) : 2차원(모든 요소들에 함수를 적용)

```python
df_last['분양가격(제곱미터)'].map(commadrop2)
# df_last['분양가격(제곱미터)'].apply(commadrop2) # 2차원일때 axis사용
```

```python
df_last['제곱미터당분양가격'] = df_last['분양가격(제곱미터)'].map(commadrop2)
df_last.tail()
```
