## 3.3 Seaborn 함수로 그래프 그리
### 1) 관계형 그래프
- scatterplot, lineplot, relplot(서브플롯제공)

```python
# 종별 다른 색상과 다른 스타일의 scatterplot
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species', palette='Set1', style='species')
plt.show()
```

```python
# 종별 다른 색상과 스타일의 lineplot
sns.lineplot(data=iris, x='petal_length', y='petal_width', hue='species', style='species', palette='Set1')
plt.show()
```

```python
# 종별 lineplot을 서브플롯으로 => relpolt(관계형 그래프에서 서브플롯 그리는 함수)
sns.relplot(data=iris, x='petal_length', y='petal_width', hue='species', 
#             kind='line', # 서브플롯으로 그려질 그래프 종류(기본값 scatter)
            col='species', # col 단위로 서브플롯 표현
#             col_wrap=2 # 한줄에 표현될 서브플롯 수
           )
plt.show()
```

```python
# ax[0]에는 산점도, ax[1]에는 lineplot
fig, ax = plt.subplots(ncols=2)
plt.subplots_adjust(wspace=0.3) # 서브플롯 사이의 공백 조정
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species', ax=ax[0])
sns.lineplot(data=iris, x='petal_length', y='petal_width', hue='species', ax=ax[1])
plt.show()
```

### 2) 범주형 그래프
- strupplot, swarmplot, barplot, barhplot, countplot, pointplot
- boxplot, violinplot, boxenplot => catplot(서브플롯) 

```python
# x축이 범주형일때의 산점도(중첩된 점)
sns.stripplot(data=iris, x='species', y='petal_width', hue='species')
# hue속성이 있을 겨우  palette가능
plt.show()
```

```python
# x축이 범주형일때의 산점도(중첩되지 않게)
sns.swarmplot(data=iris, x='species', y='petal_width', hue='species')
plt.show()
```

```python
# 내림차순 정렬된 iris로 범주형 산점도
sns.swarmplot(data=iris.sort_values('petal_width', ascending=False),
              x='species', y='petal_width', hue='species')
plt.show()
```

```python
# 산점도의 x축 순서를 임의로 정할 수 있음
sns.swarmplot(data=iris, x='species', y='petal_width', hue='species', order=['versicolor', 'virginica', 'setosa'])
plt.show()
```

```python
# barplot : 각 범주별 평균값과 그에 대한 신뢰구간
#          신뢰구간을 제외 : ci(confidence interval) = None, errorbar = None
sns.barplot(data=iris, x='species', y='sepal_length', 
            #errorbar=None
           )
sns.lineplot(data=iris, x='species', y='sepal_length')
plt.show()
```

```python
# 성별 survived의 평균을 막대그래프
sns.barplot(data=titanic, x='sex', y='survived', errorbar=None, 
            order=['female','male'],
            palette='Pastel1')
```

```python
# class별로 성별 생존률을 서브플롯(막대그래프)
sns.catplot(data=titanic, x='sex', y='survived', hue='class', 
            kind='bar', # 기본값 : stripplot
            col='class', # col단위로 서브플롯 표현
            sharey=False # y축 공유 안함
#             col_wrap=2 #  col_wrapL (기본값 1)은 한행에 모든 서브플롯 표현 
           )
```

### 5) 회귀 그래프

```python
r = iris.corr(numeric_only=True)
```

```python
sns.heatmap(r, vmin=-1, vmax=1, annot=True, fmt='.3f', cmap='Blues') # annot=True 값을 출력 ,  fmt='.3f' 소숫점 3자리
plt.show()
```

```python
# 회귀 그래프 : regplot (hue 사용불가,), lmplot(hue 사용가능)
sns.regplot(data=iris, x='petal_length', y='petal_width')
plt.show()
```

```python
sns.lmplot(data=iris, x='petal_length', y='petal_width', hue='species')
plt.show()
```
